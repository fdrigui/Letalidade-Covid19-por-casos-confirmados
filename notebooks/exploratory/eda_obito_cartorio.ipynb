```python
# Import de Bibliotecas
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# NB Configuration

%config InlineBackend.figure_format = 'retina'
sns.set_context('talk')
plt.style.use(r'../../src/visualization/my_style.mplstyle')
```

```python
path = r'../../data/raw/obito_cartorio.csv.gz'
df = pd.read_csv(path)
```

```python
sns.lineplot(x='epidemiological_week_2019', y='deaths_pneumonia_2019', data=df)
```

```python
df['deaths_pneumonia_2019_diff'] = df['deaths_pneumonia_2019'].diff()
df['deaths_pneumonia_2020_diff'] = df['deaths_pneumonia_2020'].diff()
```

```python
ax = sns.lineplot(x='epidemiological_week_2019', y='deaths_pneumonia_2019_diff',
             data=df.query('state=="SP"').groupby('epidemiological_week_2019').sum()[1:-2])
ax = sns.lineplot(x='epidemiological_week_2020', y='deaths_pneumonia_2020_diff',
             data=df.query('state=="SP"').groupby('epidemiological_week_2020').sum()[1:-2])
plt.title('Mortes por Pneumonia\nem SP no ano de 2019')
```
