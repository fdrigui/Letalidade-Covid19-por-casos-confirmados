```python
#Import de Bibliotecas

import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
%config InlineBackend.figure_format = 'retina'
sns.set_context('talk')
```

```python
path = 'https://raw.githubusercontent.com/fdrigui/polio_no_brasil_2010-2020/main/data/raw/cobertura_vacinal_brasil/cobertura_por_imuno_por_ano_1994-2021.csv'
df = pd.read_csv(path, sep=';', decimal=',', encoding='ISO-8859-1',
                                   skipfooter=2, na_values = 0.000, engine='python')
df = df.set_index('Imuno').T.drop(' Total')[['Poliomielite']].dropna()[10:]
```

```python
# Esse arquivo aqui já configura o layout do jeitão que eu gosto...
plt.style.use(r'../../src/visualization/my_style.mplstyle')
```

```python
#plt.style.use(r'../../src/visualization/my_style.mplstyle')
ax = sns.lineplot(x=df.index, y='Poliomielite', data=df, label='trts')
ax = sns.lineplot(x=df.index, y='Poliomielite', data=df*0.8, label='trts2')
ax = sns.lineplot(x=df.index, y='Poliomielite', data=df*0.6, label='trts3')
ax.set_title('Test')
plt.xlabel('Ano')
plt.show()
```
